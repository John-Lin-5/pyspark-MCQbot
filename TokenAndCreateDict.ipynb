```
def parse_interaction(line):
    elems = line.split(" ")
    tag = elems[0]
    return (tag,elems[1:])
```

```
def f(x):
  return x
```

```
import json

path = "/content/drive/MyDrive/中興課程/2022高等資料/rdd_split/wiki_rdd"

for i in range(1,10):
  print("processing sample%d" %i)
  sample_wiki_rdd = sc.textFile(path+str(i))
  testtagRDD = sample_wiki_rdd.map(parse_interaction)
  testinvRDD = testtagRDD.flatMapValues(f).map(lambda x: (x[1],x[0])).groupByKey()
  testinvIndex = testinvRDD.collect()
  my_dict = {word[0] : list(word[1]) for word in testinvIndex}
  res = {k : v for k, v in my_dict.items() if len(k) >= 2}
  res = res.items()
  with open('/content/drive/MyDrive/中興課程/2022高等資料/221123myNewDict/inv' +str(i)+ '.json', 'w', encoding='UTF-8') as fp:
      json.dump(list(res), fp, ensure_ascii=False)
```
